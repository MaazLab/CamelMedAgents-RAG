```python
import pandas as pd

# ==== File Paths ====
input_file = "/home/maaz-rafiq/work/phd_thesis_tanveer/CamelMedAgents-RAG/dataset/patient_description_sir_Suffian/PI-PD-20_consolidated_enriched.tsv"
output_file = "/home/maaz-rafiq/work/phd_thesis_tanveer/CamelMedAgents-RAG/dataset/patient_description_sir_Suffian/PI-PD-20_consolidated_enriched_with_catogries.tsv"

# ==== Load TSV ====
df = pd.read_csv(input_file, sep="\t")

# ==== Category Mapping ====
category_mapping = {
    # Ophthalmology
    "Conjunctivitis": "Ophthalmology",
    "Cataract": "Ophthalmology",

    # Cardiovascular
    "Heart Attack": "Cardiovascular",
    "Hypertension": "Cardiovascular",
    "Stroke & Tia": "Cardiovascular",

    # Gastrointestinal
    "Piles": "Gastrointestinal",
    "hernia": "Gastrointestinal",
    "Appendicitis": "Gastrointestinal",

    # Infectious Diseases
    "Dengue Fever": "Infectious Diseases",
    "Urinary Tract Infection": "Infectious Diseases",
    "Hepatitis": "Infectious Diseases",
    "Otitis Media": "Infectious Diseases",

    # Respiratory
    "Asthma": "Respiratory",
    "Cough": "Respiratory",

    # Dermatology
    "acne": "Dermatology",
    "Urticarial Rash": "Dermatology",

    # Neurology
    "Headache": "Neurology",
    "Alzheimer": "Neurology",

    # Hematology / Nutritional
    "B12 Deficiency": "Hematology / Nutritional",

    # Oncology
    "Cancer": "Oncology",
}

# ==== Add Category Column ====
df["Category"] = df["Label"].map(category_mapping)

# If any labels were not mapped, mark them
df["Category"] = df["Category"].fillna("Other")

# ==== Move Category Column to Start ====
cols = df.columns.tolist()
cols.insert(0, cols.pop(cols.index("Category")))
df = df[cols]

# ==== Save New File ====
df.to_csv(output_file, sep="\t", index=False)

print("Done ✅ File saved as:", output_file)
```
